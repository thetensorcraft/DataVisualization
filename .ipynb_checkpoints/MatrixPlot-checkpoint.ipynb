```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**HEAT MAP**

```python
# load dataset
flight = sns.load_dataset('flights')
flight
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# find the correlation
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
# plot the heatmap
sns.heatmap(tipscorr.corr(), annot = True)
```
