```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**MATPLOTLIB BASICS**

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
# linegraph
plt.plot(x, y)
plt.title('Axis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

**SUBPLOTS IN MATPLOTLIB**

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.subplot(2, 2, 4)
plt.plot(x, y)
```

**OBJECT ORIENTED WAY**

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x, y)
axis2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axis2.plot(x, y)
```

**TYPES OF PLOTS**

```python
# scatter plot
plt.scatter(x, y)
```

```python
# histogram plot
plt.hist(x)
```

```python
# box plot
plt.boxplot(x)
```
